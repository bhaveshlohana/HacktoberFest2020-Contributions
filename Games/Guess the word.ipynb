```python
import random
name = input("Enter your name: ")
print(f"Hello! Welcome {name}")

words = ["apple","banana","cough","donkey","enemies","fellow","golden","hinder","intestine","jumble","kernel","lotion"]

word = random.choice(words)
# word = "apple"
guessed = ""
counter=10

preguess = random.sample(list(word),random.randint(0,2))
for i in preguess:
    if(i not in guessed):
        guessed += i

while(counter>0):
    w = True
    for i in word:
        if(i in guessed):
            print(i,end=" ")
        else:
            print("_",end=" ")
            w = False
    print()
    print("--------------------------------------------------")
    if(w):
        print("Congratulations!!!! You won")
        print("--------------------------------------------------")
        break
    
    ch = input("Enter a character: ")
    ch = ch.lower()
    if(len(ch)!=1):
        print("You need to enter only 1 character")
    else:
        if(ch in guessed):
            print("You've already guessed this character")
            counter -= 1
        elif(ch in word):
            print("Correct guess")
            guessed += ch
        else:
            print("Wrong guess")
            guessed += ch
            counter -= 1
        print(f"You have {counter} chances remaining")
        if(counter == 0):
            print("--------------------------------------------------")
            print("You lost!!")
            print("--------------------------------------------------")
            print(f"Correct word was {word}")
            print("--------------------------------------------------")
            break


```
