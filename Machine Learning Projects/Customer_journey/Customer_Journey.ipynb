```python
def event_channel(cols):
    event = cols[1]
    channel = cols[0]
    if event=='enquiry' and channel=='web':
        return 0
    if event=='quote' and channel=='web':
        return 1
    if event=='policy_buy' and channel=='web':
        return 2
    if event=='policy_change' and channel=='web':
        return 3
    if event=='policy_cancel' and channel=='web':
        return 4
    if event=='claim' and channel=='web':
        return 5
    if event=='Renew' and channel=='web':
        return 6
    if event=='enquiry' and channel=='mobile':
        return 7
    if event=='quote' and channel=='mobile':
        return 8
    if event=='policy_buy' and channel=='mobile':
        return 9
    if event=='policy_change' and channel=='mobile':
        return 10
    if event=='policy_cancel' and channel=='mobile':
        return 11
    if event=='claim' and channel=='mobile':
        return 12
    if event=='Renew' and channel=='mobile':
        return 13
    if event=='enquiry' and channel=='ccenter':
        return 14
    if event=='quote' and channel=='ccenter':
        return 15
    if event=='policy_buy' and channel=='ccenter':
        return 16
    if event=='policy_change' and channel=='ccenter':
        return 17
    if event=='policy_cancel' and channel=='ccenter':
        return 18
    if event=='claim' and channel=='ccenter':
        return 19
    if event=='Renew' and channel=='ccenter':
        return 20
    
```

```python
action1=data[['Channel_Type1','Event_type1']].apply(event_channel,axis=1)
action1
```

```python
action2=data[['Channel_Type2','Event_type2']].apply(event_channel,axis=1)
action2.head()
```

```python
action3=data[['Channel_Type3','Event_type3']].apply(event_channel,axis=1)

action3.head()
```

```python
action_predict=data[['Channel_Type_predict','Event_type_predict']].apply(event_channel,axis=1)
action_predict.head()
```

```python
sex_df = pd.get_dummies(data['sex'],drop_first=3)
sex_df.head()
```

```python
new_df = pd.get_dummies(data['New Customer'],drop_first=3)
new_df.head()
```

```python
data.drop(['Channel_Type1','Customer_id','Event_type1','Channel_Type2','Channel_Type3','Event_type2','Event_type3','Event_type_predict','Channel_Type_predict','visitor_id','visitor_location','sex','visitor_duration','New Customer'], axis = 1,inplace=True)
```

```python
data.head()
```

```python
data = pd.concat([data,sex_df,new_df,action1,action2,action3,action_predict],axis=1)
```

```python
data.head()
```

```python
data=data.rename(columns={0: 'Action1',1: 'Action2',2: 'Action3',3: 'Action_predict','M':'Sex','visitor_date':'Total days','yes':"New Customer"})
data.head()
```

```python
data['Total days'] = data['Total days'].apply(lambda x: x.strip(" ")[0])
data['Total days']=data['Total days'].astype('int')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('Action_predict',axis=1), 
                                                    data['Action_predict'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier = classifier.fit(X_train,y_train)
```

```python
#prediction
y_pred = classifier.predict(X_test)
#Accuracy
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_test,y_pred))
```

```python
classifier1 = DecisionTreeClassifier(criterion="entropy", max_depth=5)
# Train Decision Tree Classifer
classifier1 = classifier1.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = classifier1.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
