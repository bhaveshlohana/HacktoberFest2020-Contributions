# K-Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Loading

```python
PATH = "../../../Clustering/KMeans_Clustering/Python/Mall_Customers.csv"
```

```python
dataset = pd.read_csv(PATH)
```

```python
X = dataset.iloc[:, [3,4]].values
```

## Elbow Method

```python
from sklearn.cluster import KMeans

number_of_clusters = 10
wcss = []
plt.figure(figsize=(12, 8))
for i in range(1, number_of_clusters + 1):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, number_of_clusters + 1), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

## K-Means

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, init="k-means++", random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualisation

```python
plt.figure(figsize=(12, 8))

for i in range(n_clusters):
    label = "Cluster " + str(i + 1)
    plt.scatter(X[(y_kmeans == i), 0], X[(y_kmeans == i), 1], s=100, label=label)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=500, c='yellow')
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.legend()
plt.show()
```
