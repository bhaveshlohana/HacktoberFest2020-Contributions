# Activation Function
# WAP to plot activation function sigmoid, relu, threshold and linear.
# threshold: Y=f(x)=0 if x<=0, else y=1.
# linear Y=F(x)=x
# relu: Y=f(x)=0 if x<0, else y=x
# sigmoid: y=f(x)=1/(1+exp(-x))


```python
import numpy as np
import tensorflow as tf
x=tf.constant(np.arange(-10,10,0.01))
y_sigmoid=tf.nn.sigmoid(x)
y_linear=x
y_relu=tf.nn.relu(x)
y_tanh=tf.nn.tanh(x)

import matplotlib.pyplot as plt
plt.plot(x,y_sigmoid,label='sigmoid')
plt.plot(x,y_relu,label='relu')
plt.plot(x,y_tanh,label='tanh')
plt.plot(x,y_linear,label='linear')
plt.legend()
plt.xlabel('x-->')
plt.ylabel('Y->')
plt.title('Activation Function')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
x=tf.constant(np.arange(-10,10,0.01))
y_sigmoid=tf.nn.sigmoid(x)
y_linear=x
y_relu=tf.nn.relu(x)
y_tanh=tf.nn.tanh(x)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.subplot(3,3,1)
plt.plot(x,y_sigmoid,label='sigmoid')
plt.xlabel('x-->')
plt.ylabel('Y->')
plt.title('Activation Function')

plt.subplot(3,3,4)
plt.plot(x,y_relu,label='relu')
plt.xlabel('x-->')
plt.ylabel('Y->')
plt.title('Activation Function')

plt.subplot(3,3,7)
plt.plot(x,y_tanh,label='tanh')
plt.subplot(3,3,9)
plt.plot(x,y_linear,label='linear')
#plt.legend()
plt.xlabel('x-->')
plt.ylabel('Y->')
plt.title('Activation Function')
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
model=keras.Sequential(
[
    layers.Dense(2,activation='relu',input_shape=(3,)),
    layers.Dense(3,activation='relu'),
    layers.Dense(4,activation='relu')
    
])
print(model)
print(model.layers)
x=tf.ones((4,3))
y=model(x)
print(model.summary())
keras.utils.plot_model(model,'model1.png',show_shapes=True)

```
