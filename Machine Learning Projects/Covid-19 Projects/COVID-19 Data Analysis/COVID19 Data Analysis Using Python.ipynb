## The following notebook is the EDA of Covid-19 along with the correlation with the various life factors to see if there is any relationship between the spread of the virus in a country and how happy people are, living in that country.

##### The COVID19 dataset, published by John Hopkins University, which consist of the data related to cumulative number of confirmed cases, per day, in each Country and the second dataset-wolrd happiness report Dataset consist of various life factors, scored by the people living in each country around the globe.

```python
#importing modules

import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#importing covid19 the dataset

df = pd.read_csv('Datasets\covid19_Confirmed_dataset.csv')
df.head(10)
```

```python
#Dropping Useless Columns
df.drop(['Lat','Long'],axis=1,inplace=True)
df.head()
```

```python
#Aggregating the rows by the country

df = df.groupby("Country/Region").sum()
df.head()
```

```python
df.loc['China'].plot()
df.loc['India'].plot()
df.loc['US'].plot()
plt.legend()
```

```python
df.loc['India'][:70].plot()
```

```python
df.loc['India'].diff().plot() #rate of the spread
```

```python
df.loc['India'].diff().max()
```

```python
df.loc['US'].diff().max()
```

```python
#maximum infection rate

countries = list(df.index)
max_infection_rates = []
for country in countries :
    max_infection_rates.append(df.loc[country].diff().max())
df['max infection rate'] = max_infection_rates
```

```python
df.head()
```

```python
data = pd.DataFrame(df['max infection rate'])
data.head()
```

```python
world_happiness_report = pd.read_csv("Datasets\worldwide_happiness_report.csv")
world_happiness_report.head()
```

```python
#dropping the useless columns

columns_to_dropped = ['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
world_happiness_report.head()
```

```python
#changing the indices of the dataframe

world_happiness_report.set_index(['Country or region'],inplace=True)
world_happiness_report.head()
```

```python
df_final = world_happiness_report.join(data).copy()
df_final.head()
```

```python
df_final.corr() # it is representing the currelation between every two columns of our dataset 
```

```python
#Plotting GDP vs maximum Infection rate

x = df_final['GDP per capita']
y = df_final['max infection rate']
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))
```

```python
#Plotting Social support vs maximum Infection rate

x = df_final['Social support']
y = df_final['max infection rate']
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))
```

```python
#Plotting Healthy life expectancy vs maximum Infection rate

x = df_final['Healthy life expectancy']
y = df_final['max infection rate']
sns.scatterplot(x,np.log(y))
```
