### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Open Image

```python
img=cv2.imread('Images/car_plate.jpg')
```

### Convert Image

```python
def display(img):
    fig=plt.figure(figsize=(10,8))
    ax=fig.add_subplot(111)
    new_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

### Display it

```python
display(img)
```

### Create Classifier

```python
plate_classifier=cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### It's a Kind of Magic

```python
def detect_plate(img):
    plate_img=img.copy()
    plate_rects = plate_classifier.detectMultiScale(plate_img, 1.1, 1)
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img,(x,y),(x+w,y+h),(255,255,2555),5)
    return plate_img
```

### Check the Results

```python
result=detect_plate(img)
```

```python
display(result)
```
